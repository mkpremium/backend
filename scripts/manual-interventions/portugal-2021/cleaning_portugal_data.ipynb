```python
raw = pd.read_excel('BBDD_CARDERNETAS_CORREGIDO.xlsx', sheet_name="sin duplicados", dtype=str)
print(raw.columns)
print(raw.info())
```

```python
arr = raw['CALLE'].str.strip().str.split(' ').str[0].sort_values().unique()
df = pd.DataFrame(data=arr, index=None, columns=['Prefijo'])
df.to_excel('prefijo_calles.xlsx', index=None)
```

```python
raw['CALLE'].str.strip(" -.\n").str.split(' ').str[0].value_counts().sort_index().to_excel('prefijo_calles.xlsx', index_label=["Calle"])
```

```python
raw['STREET_PREFIX'] = raw['CALLE'].str.strip(" -.\n").str.split(' ').str[0]
```

```python
raw['TIPO_CLEAN'] = raw['TIPO'].str.strip(' .').str.upper()
raw['TIPO_CLEAN']
```

```python
# raw['TIPO_CLEAN'].sort_index().value_counts().to_excel('tipos_via.xlsx')
raw['TIPO_CLEAN'].value_counts(sort=False).sort_index().to_excel('tipos_via.xlsx')
```

```python
raw['SLUG'] = raw['AM'].map(str) + '-' + raw['A'].map(str) + '-' + raw['TIPO_CLEAN'].map(str) + '-' + raw['CALLE'].map(str).str.strip() + '-' + raw['Nº'].map(str) + '-' + raw['BARRIO'].map(str)+ '-' + raw['CIUDAD'].map(str)
# raw['SLUG'] = raw['SLUG'].replace(to_replace=r'[\s\.,ºª]', value='', regex=True)
raw['SLUG'] = raw['SLUG'].replace(regex={r'[\.,ºª]': '', r'\s': '_'})
print(raw['SLUG'].size)
print(raw['SLUG'].unique().size)
```

```python
import uuid
import random

def random_uuid():
    return uuid.UUID(bytes=bytes(random.getrandbits(8) for _ in range(16)), version=4)

random.seed('portugal-data')
raw['BUILDING_ID'] = [random_uuid() for _ in range(len(raw.index))]
raw['BUILDING_ID']
```

```python
raw.to_csv('portugal_ready.csv', sep=";")
```
